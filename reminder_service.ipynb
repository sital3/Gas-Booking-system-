```python
# Reminder scheduler (run this once in a notebook cell before launching GUI)
from apscheduler.schedulers.background import BackgroundScheduler
from datetime import datetime
import tkinter.messagebox as msg

scheduler = BackgroundScheduler()
scheduler.start()

def show_popup_reminder(name, date):
    msg.showinfo("Booking Reminder", f"Reminder: LPG booking for {name} on {date}!")

def schedule_gui_reminder(remind_at, name, date):
    """
    remind_at: datetime obj (when to show)
    name, date: strings used in popup
    """
    if remind_at <= datetime.now():
        print("Reminder time is in the past. Skipping.")
        return
    scheduler.add_job(
        show_popup_reminder,
        'date',
        run_date=remind_at,
        args=[name, date]
    )
    print("Reminder scheduled for:", remind_at)
```
